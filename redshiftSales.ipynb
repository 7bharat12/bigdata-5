```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
dataset = pd.read_csv(r'C:\Users\bhara\Desktop\BharatChandra_AWSBigData5\sales.csv')

# Perform analysis and visualization

# Bar chart - Sales by Genre
sales_by_genre = dataset.groupby('genre')['global_sales'].sum()

plt.bar(sales_by_genre.index, sales_by_genre.values)
plt.xlabel('Genre')
plt.ylabel('Total Sales')
plt.title('Sales by Genre')
plt.xticks(rotation=45)
plt.show()

# Pie chart - Sales by Platform
sales_by_platform = dataset.groupby('platform')['global_sales'].sum()
plt.figure(figsize=(8, 8)) 
plt.pie(sales_by_platform.values, labels=sales_by_platform.index, autopct='%1.1f%%')
plt.axis('equal')
plt.title('Sales by Platform')
plt.show()

# Line chart - Sales over Years
sales_over_years = dataset.groupby('year')['global_sales'].sum()

plt.plot(sales_over_years.index, sales_over_years.values)
plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Sales over Years')
plt.show()

# Scatter plot - Sales in North America vs. Europe
plt.figure(figsize=(8, 6))
plt.scatter(dataset['na_sales'], dataset['eu_sales'])
plt.xlabel('Sales in North America')
plt.ylabel('Sales in Europe')
plt.title('Sales in North America vs. Europe')
plt.show()

```
